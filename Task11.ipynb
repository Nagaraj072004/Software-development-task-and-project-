```python

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Data: Year and Population
year = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]).reshape((-1, 1))
population = np.array([1000000, 1100000, 1200000, 1300000, 1400000, 1500000, 1600000, 1700000, 1800000, 1900000])

# Create and fit the model
model = LinearRegression()
model.fit(year, population)

# Predict population for the year 2025
year_to_predict = np.array([[2025]])
predicted_population = model.predict(year_to_predict)

# Print results
print(f"Predicted population for the year 2025: {predicted_population[0]:.2f}")

# Plot the data and regression line
plt.scatter(year, population, label="Actual population")
plt.plot(year, model.predict(year), color="red", label="Regression line")
plt.scatter(year_to_predict, predicted_population, color="green", label="Predicted population")
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Prediction")
plt.legend()
plt.show()
```
