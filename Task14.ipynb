```python
#matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.logspace(0, 3, 100)
y = np.log(x)

# Create the plot
plt.plot(x, y)
plt.xscale('log')
plt.yscale('log')
plt.title('Logarithmic Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True, which="both", axis="both", ls="--")
plt.show()
```
